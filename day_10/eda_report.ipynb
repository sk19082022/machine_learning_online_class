### EDA FOR YOUTUBE VIDEO IN 2020 AUGUST

```python
import pandas as pd
import matplotlib.pyplot as plt
yt_dataset = pd.read_csv("20.22.08_IN_videos.csv")

df = pd.DataFrame(yt_dataset)
print(df.head(10))
```

## 1. What are the Top 10 Most Viewed Videos?

```python
top_views = df.sort_values(
    by='view_count',
    ascending=False,
)[['title', 'view_count']].head(10)

plt.barh(top_views['title'],top_views['view_count'])
plt.title("top views videos")
plt.xlabel('views')
plt.ylabel('title of the video')
plt.show()
```

## 2. What are the Top 10 Most Liked Videos?

```python
top_likes = df.sort_values(
    by='likes',
    ascending=False,
)[['title', 'likes']].head(10)

plt.barh(top_likes['title'],top_likes['likes'])
plt.title("top 10 liked videos")
```

## 3. Which Videos Received the Most Comments?

```python
top_comments = df.sort_values(
    by = "comment_count",
    ascending=False
)[['title','comment_count']].head(10)

plt.barh(top_comments['title'],top_comments['comment_count'])
```

## 4. Which Videos Received the Most Dislikes?

```python
most_dislikes = df.sort_values(
    by = "dislikes",
    ascending=False
)[['title','dislikes']].head(10)

plt.barh(most_dislikes['title'],most_dislikes['dislikes'])
```

## 5. What is the Distribution of Views?

```python
plt.hist(df['view_count'], bins=20)

plt.title("Distribution of Views")
plt.xlabel("Views")
plt.ylabel("Frequency")

plt.show()
```

## 6. Is There a Relationship Between Views and Likes?

```python
import seaborn as sns

sns.scatterplot(
    data = df,
    x='view_count',
    y = 'likes',
)

plt.xlabel('views')
plt.ylabel('likes')
plt.show()

# Do videos with more views generally receive more likes?
```

## 10. Which Videos Have the Highest Engagement?

```python
#create an engagement 
df['engagement'] = (
    df['likes']+df['comment_count']+df['view_count']
)

#create a top dataset
top_engagement = df.sort_values(
    by='engagement',
    ascending=False
)[['title','engagement']].head(10)

plt.barh(top_engagement['title'],top_engagement['engagement'])
plt.xlabel('engagement')
plt.ylabel('title')
plt.show()
```
