Exploratory Data Analysis (EDA)

CSV file creation and use

```python
import pandas as pd

data = pd.read_csv('data.csv')
print(data)
print("\n\n\nDescribe data : \n\n")
print(data.describe())
print("\n\nhead by one data : \n\n")
print(data.head(1))
print("\n\n\nTail by one data : \n\n")
print(data.tail(1))
print("\n\n\nData information : \n\n")
print(data.info())
```

Check null values

```python
import pandas as pd
data = pd.read_csv('data.csv')
print("\n\n\nIs any null values\n\n")
print(data.isnull())
```

Duplicate values check in csv file

```python
import pandas as pd 
data = pd.read_csv('data.csv')
print("\n\n\nIs any duplicated values\n\n")
print(data.duplicated())
```
