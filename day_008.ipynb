## Python seaborn

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from  sklearn.datasets import load_iris
import matplotlib.pyplot as plt
#iris is predifined dataset with flowers
iris =  load_iris()
df = pd.DataFrame(
    data = iris.data,
    columns = iris.feature_names
)

print(df)
```

# seaborn scatter plot

```python
sns.scatterplot(
    x = 'sepal length (cm)',
    y = 'sepal width (cm)',
    data = df
)
plt.title("sapal length vs width")
plt.xlabel('length')
plt.ylabel('width')
plt.show()
```

# With box plot

```python

sns.boxplot(
    x = 'petal length (cm)',
    y = 'petal width (cm)',
    data = df
)

plt.title("petal length vs width by santhu")
plt.xlabel('length')
plt.ylabel('width')
plt.show()
```

# Voilin plot

```python
sns.violinplot(
    x = 'sepal length (cm)',
    data = df
)

plt.title("sapal length vs width by voiline")
plt.xlabel('length')
plt.ylabel('width')
plt.show()
```

# Swarm plot

```python
sns.swarmplot(
    x = 'sepal length (cm)',
    y = 'sepal width (cm)',
    data = df   
)
plt.title("sapal length vs width by swarm")
plt.xlabel('length')
plt.ylabel('width')
plt.show()
```

# Bar plot

```python
sns.barplot(
    x = 'sepal length (cm)',
    y = 'sepal width (cm)',
    data = df      
)

plt.title("sapal length vs width by bar")
plt.xlabel('length')
plt.ylabel('width')
plt.show()
```

# Point plot

```python
sns.pointplot(
    x = 'sepal length (cm)',
    y = 'sepal width (cm)',
    data = df
)
plt.title("sapal length vs width by point")
plt.xlabel('length')    
plt.ylabel('width')
plt.show()
```

# Kenrnal desity estimation 

```python
sns.kdeplot(
    x = 'sepal length (cm)',
    y = 'sepal width (cm)',
    data = df

)

plt.title("sapal length vs width by kde")
plt.xlabel('length')
plt.ylabel('width')
plt.show()
```

# Changing Styles and Grids using sns.set_style()

```python
sns.set_style('whitegrid')
sns.pointplot(
    x = 'sepal length (cm)',
    y = 'sepal width (cm)',
    data = df
)
plt.title("sapal length vs width with white grid")
plt.xlabel('length')
plt.ylabel('width') 
plt.show()
```

```python
sns.set_style('darkgrid')
sns.pointplot(
    x = 'sepal length (cm)',
    y = 'sepal width (cm)',
    data = df
)
plt.title("sapal length vs width with white grid")
plt.xlabel('length')
plt.ylabel('width') 
plt.show()
```

```python
sns.set_style('dark')
sns.pointplot(
    x = 'sepal length (cm)',
    y = 'sepal width (cm)',
    data = df
)
plt.title("sapal length vs width with white grid")
plt.xlabel('length')
plt.ylabel('width') 
plt.show()
```

```python
sns.set_style('ticks')
sns.pointplot(
    x = 'sepal length (cm)',
    y = 'sepal width (cm)',
    data = df
)
plt.title("sapal length vs width with white grid")
plt.xlabel('length')
plt.ylabel('width') 
plt.show()
```
