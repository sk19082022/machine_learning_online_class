LINE PLOT 

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [1,2,3,4,5]

plt.plot(x,y)
plt.xlabel("x axis")
plt.ylabel("y asis")
plt.title("Line plot by santhu")
plt.show()
```

BAR GRAPH

```python
import matplotlib.pyplot as plt

name = ["A","B","C","D","E"]
value = [10,20,30,40,50]

plt.bar(name,value)
plt.xlabel("Name")
plt.ylabel("Value")
plt.title("Bar graph by santhu")
plt.show()
```

PIE CHART

```python
import matplotlib.pyplot as plt

name = ["A","B","C","D","E"]
value = [10,10,10,10,10]

plt.pie(value,labels=name)
plt.title("pie chart by santhu")
plt.show()
```

SCATTER PLOTS

```python
import matplotlib.pyplot as plt

v1 = [1,2,3,4,5]
v2 = [10,20,30,40,50]

plt.scatter(v1,v2)
plt.xlabel("v1")
plt.ylabel("v2")
plt.title("scatter plot by santhu")
plt.show()
```

Histogram

```python
import matplotlib.pyplot as plt

something = [10,20,30,40,50]

plt.hist(something,bins = 4)
plt.xlabel("something")
plt.ylabel("increasing")
plt.title("histogram by santhu")
plt.show()
```
